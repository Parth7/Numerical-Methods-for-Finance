```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Defining the parameters
mu = 2/3;
nsample = 1000000;
x = np.arange(0,5,0.05);

# Computing the PDF and CDF
f1 = (1/mu)*np.exp(-x/mu);
F1 = 1-np.exp(-x/mu);

#Compare with inbuilt PDF
f = expon.pdf(x,loc = 0, scale = 1);
F = expon.cdf(x,loc = 0, scale = 1);

# Create Figure 
fig = plt.figure(figsize=(7,4), dpi=100)

# Add set of axes to figure
ax = fig.add_axes([0,0,1,1]) # left, bottom, width, height 

#Figure 1
#Plot on that set of axes
ax.plot(x,f1,'r', label = 'PDF')
ax.plot(x,F,'g', label  = 'CDF')
ax.plot(x,f,'b', label = 'PDF')
plt.legend()
ax.set_title('Exponential distribution with mu = 2/3')

#Figure 2
s = expon.rvs(size = nsample);
plt.hist(s,100,density=True, label = 'sampled', alpha=0.5)
plt.xlim(0,5);
#ax.plot(x, f1, 'r', label = 'PDF')
plt.legend()
plt.show()
```
