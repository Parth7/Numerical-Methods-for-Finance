```python
#!/usr/bin/python
#Assignment2_2.ipynb
#Parth Parakh
#purpose: Formative Assignment 2_2

#---------------------------------------------------------------------
#import modules
import numpy as np
import matplotlib.pyplot as plt
#---------------------------------------------------------------------
#Initialise values
ngrid = 200;
nsample = 10000;

# (a) Compute the PDF
dx = 1/ngrid;
x = np.arange(0,1,dx);
f = 2/(np.pi*np.sqrt(1-x**2));

# (b) Sample
dy = np.pi/ngrid;
y = np.arange(-np.pi/2,np.pi/2,dy);
U = np.pi*(np.random.rand(nsample,1)-1/2);
hu,size1 = np.histogram(U,y)/(nsample*(y[1]-y[0]));
#print(hu)
X = np.cos(U);
hx,size2 = np.histogram(X,x)/(nsample*(x[1]-x[0]));

#Figure 1
# Plot the PDF of U
plot1 = plt.figure(1,figsize=(7,4), dpi=100)
plt.plot(y[:-1],hu,'b', label = 'Sampled')
plt.plot(y,np.ones(len(y))/np.pi, label = 'Theory')
plt.ylim(-0.2,0.8)
plt.xlabel('x')
plt.ylabel('f')
plt.title('Distribution of U')
plt.legend()

#Figure 2
# Plot the PDF of X
plot2 = plt.figure(2,figsize=(7,4), dpi=100)
plt.plot(x[:-1],hx,'r', label = 'Sampled')
plt.plot(x,f,'b',label = 'Theory')
plt.xlabel('x')
plt.ylabel('f')
plt.title('Distribution of X = cos U')
plt.legend()

plt.show()
```
