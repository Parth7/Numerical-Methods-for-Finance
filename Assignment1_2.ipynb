```python
#!/usr/bin/python
#Assignment1_2.ipynb
#Parth Parakh
#purpose: Formative Assignment 1_2

#---------------------------------------------------------------------
#import modules
import numpy as np
import matplotlib.pyplot as plt
import time
#---------------------------------------------------------------------
def inbuilt(n):
    l = n*n;
    np.random.seed(2)
    A = np.random.rand(l).reshape(n,n)
    B = np.random.rand(l).reshape(n,n)

    #Compute the matrix product C = A*B (a)
    start_time = time.process_time()
    C_a = np.matmul(A,B);
    #print(C_a)
    pt = time.process_time() - start_time;
    return pt;
#---------------------------------------------------------------------
def threeloops(n):
    l = n*n;
    np.random.seed(2)
    A = np.random.rand(l).reshape(n,n)
    B = np.random.rand(l).reshape(n,n)

    #Compute the matrix product C = A*B (a)
    start_time = time.process_time()
    C_b = np.zeros(l).reshape(n,n);
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C_b[i,j] = C_b[i,j] + A[i,k]*B[k,j];
    pt = time.process_time() - start_time;
    return pt;
#---------------------------------------------------------------------
def matm(n):
    l = n*n;
    np.random.seed(2)
    A = np.random.rand(l).reshape(n,n)
    B = np.random.rand(l).reshape(n,n)

    #Compute the matrix product C = A*B (a)
    start_time = time.process_time()
    C_a = np.matmul(A,B);
    #print(C_a)
    print (time.process_time() - start_time, "seconds")
    
    #Compute matrix product using 3 for loops (b)
    start_time = time.process_time()
    C_b = np.zeros(l).reshape(n,n);
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C_b[i,j] = C_b[i,j] + A[i,k]*B[k,j];
    #print(C_b)
    print (time.process_time() - start_time, "seconds")

    #Compute matrix product using 3 for loops i and j inverse (c)
    start_time = time.process_time()
    C_c = np.zeros(l).reshape(n,n);
    for j in range(n):
        for i in range(n):
            for k in range(n):
                C_c[i,j] = C_b[i,j] + A[i,k]*B[k,j];
    #print(C_c)
    print (time.process_time() - start_time, "seconds")

    #Compute matrix product using 2 for loops i and j and sum (d)
    start_time = time.process_time()
    C_d = np.zeros(l).reshape(n,n);
    for i in range(n):
        for j in range(n):
            C_d[i,j] = sum (A[i,:]*B[:,j]);
    #print(C_d)
    print (time.process_time() - start_time, "seconds")

    #Compute matrix product using 2 for loops i and j and transpose A (e)
    start_time = time.process_time()
    A_t = A.transpose()
    C_e = np.zeros(l).reshape(n,n);
    for i in range(n):
        A_tt = A_t[:,i].transpose();
        for j in range(n):
            C_e[i,j] = sum(A_tt*B[:,j]);
    #print(C_e)
    print (time.process_time() - start_time, "seconds")
#----------------------------------------------------------------------
# main caller
#Compute a to e and note time in secs (f)
number = [3,10,100,1000];
for n in number:
    matm(n);
    
#Compute and plot log-log graph (g)
number1 = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000];
l1 = len(number1);
time1 = np.zeros(l1);
for i in range(l1):
    time1[i] = inbuilt(number1[i]);

#print(time1)
plt.loglog(number1,time1,'ro',label='Inbuilt');
    
number2 = [ 10, 20, 50, 100, 200, 500, 1000];
l2 = len(number2);
time2 = np.zeros(l2);
for i in range(l2):
    time2[i] = threeloops(number2[i]);

#print(time2)
plt.loglog(number2,time2,'bo',label='3loops');

#Figure properties
plt.legend()
plt.show()
```
