```python
#!/usr/bin/python3.7
#GBM.ipynb
#Parth Parakh
#purpose: Simulate Geometric Brownian Motion
#ds = s*mu*dt + s*sigma*dW(t)

#---------------------------------------------------------------------
#import modules
import numpy as np
import matplotlib.pyplot as plt
#----------------------------------------------------------------------

#Defining the parameters
npaths = 20000;           #number of paths
T = 1;                   #time horizon
nsteps = 200;            #number of time steps
dt = T/nsteps;           #time step
t = np.arange(0,T,dt);   #observation times
mu = 0.2; sigma = 0.4; #model parameters
S0 = 1;

drift = (mu-0.5*(sigma**2))*dt;
diffusion = sigma*np.sqrt(dt)*np.random.normal(0,1,(npaths,nsteps));
dx = drift + diffusion;
x = np.cumsum(dx,axis=1);
S = S0*np.exp(x);
Ex = np.exp(mu*t);

#----------------------------------------------------------------------

plot1 = plt.figure(1,figsize=(7,4), dpi=100)
for i in range(0,100):
    plt.plot(t,S[i],alpha=0.5)
plt.plot(t,Ex,'black', label = 'Mean path')
plt.ylim(0,2.5)
plt.xlim(0,1)
plt.legend()
plt.xlabel('t')
plt.ylabel('x')
plt.title('Paths of an Geometric Brownian motion ds = s*mu*dt + s*sigma*dW(t)')
plt.show()

plot2 = plt.figure(2,figsize=(7,4), dpi=100)
s = np.arange(0,3.5,0.035);
plt.subplot(3,1,1)
plt.hist(S[:,20],s)
plt.xlim(0,3.5)
plt.tight_layout()

plt.subplot(3,1,2)
plt.hist(S[:,80],s)
plt.xlim(0,3.5)
plt.tight_layout()

plt.subplot(3,1,3)
plt.hist(S[:,199],s)
plt.xlim(0,3.5)
plt.tight_layout()
```
