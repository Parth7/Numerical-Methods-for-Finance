```python
#!/usr/bin/python
#Assignment2_1.ipynb
#Parth Parakh
#purpose: Formative Assignment 2_1

#---------------------------------------------------------------------
#import modules
import numpy as np
import matplotlib.pyplot as plt
#---------------------------------------------------------------------
#Initialise values
a = 1597; 
b = 51749; 
M = 244944;
seed = 12345;
seed = np.mod(seed,M);
nsample = 10000;

#Generator
U = np.zeros(nsample);
U[0] = seed;
for i in range(1,nsample):
    U[i] = np.mod(a*U[i-1]+b,M);  
U = U/M;

#Histogram
nbins = 100;
h,x = np.histogram(U,nbins);
h = h*nbins/nsample;

#Figure 1
plot1 = plt.figure(1,figsize=(7,4), dpi=100)
#ax = fig.add_axes([0,0,1,1])
plt.plot(U[:-1],U[1:],'r,', label = 'Random Numbers')
plt.legend()

#Figure 2
plot2 = plt.figure(2,figsize=(7,4), dpi=100)
#ax = fig.add_axes([0,0,1,1])
plt.plot(x[:-1],h,'b', label = 'Sampled')
plt.plot(x[:-1],np.ones((int)(nbins)), label = 'Theory')
plt.ylim(0,2)
plt.xlim(-.20,1.20)
plt.legend()

plt.show()
```
