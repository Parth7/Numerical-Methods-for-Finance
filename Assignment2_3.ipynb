```python
#!/usr/bin/python
#Assignment2_3.ipynb
#Parth Parakh
#purpose: Formative Assignment 2_3

#---------------------------------------------------------------------
#import modules
import numpy as np
import matplotlib.pyplot as plt
#---------------------------------------------------------------------
#Initialise values
a = 1366; 
b = 150889; 
M = 714025;
seed = 12345;
seed = np.mod(seed,M);
nsample = 10000;

#Congruential Generator
U = np.zeros(nsample);
U[0] = seed;
for i in range(1,nsample):
    U[i] = np.mod(a*U[i-1]+b,M);  

#Fibonnaci Generator
nu = 17;
mu = 5;
#Fibonacci generator
for i in range(nu+1,nsample):
    U[i] = np.mod(U[i-nu]-U[i-mu],M);  
U = U/M;

# Bin the random variables in a histogram and normalise it
nbins = 100;
h,x = np.histogram(U,nbins);
h = h*nbins/nsample; # normalisation

#Figure 1
plot1 = plt.figure(1,figsize=(7,4), dpi=100)
#ax = fig.add_axes([0,0,1,1])
plt.plot(U[:-1],U[1:],'r,', label = 'Scatter plot')
plt.legend()

#Figure 2
plot2 = plt.figure(2,figsize=(7,4), dpi=100)
#ax = fig.add_axes([0,0,1,1])
plt.plot(x[:-1],h,'b', label = 'Sampled')
plt.plot(x[:-1],np.ones((int)(nbins)), label = 'Theory')
plt.ylim(0,2)
plt.xlim(-.20,1.20)
plt.legend()
plt.show()
```
