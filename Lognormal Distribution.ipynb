```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Defining the parameters
mu = np.exp(0.2);             #mean
sigma = 0.1;          #standard deviation
nsample = 1000000;

x = np.arange(0,2,0.02);

#Compare with inbuilt PDF
f = lognorm.pdf(x,s=sigma,loc=0,scale=mu);
F = lognorm.cdf(x,s=sigma,loc=0,scale=mu);

# Create Figure 
fig = plt.figure(figsize=(7,4), dpi=100)

# Add set of axes to figure
ax = fig.add_axes([0,0,1,1]) # left, bottom, width, height 

#Figure 1
#Plot on that set of axes
ax.plot(x,F,'r', label  = 'CDF')
ax.plot(x,f,'b', label = 'PDF')
plt.legend()
ax.set_title('Lognormal distribution with mu = 0.2 and sigma = 0.1')

#Figure 2
X = np.random.lognormal(0.2,sigma,nsample);
#y = np.exp(X);
plt.hist(X,100,density=True, label = 'sampled', alpha=0.5)
#ax.plot(x, f1, 'r', label = 'PDF')
plt.xlim(0,2)
plt.legend()
plt.show()
```
