```python
#!/usr/bin/python3.7.4
#Coursework1_1.py

#---------------------------------------------------------------------
#import modules
import numpy as np
import matplotlib.pyplot as plt
#---------------------------------------------------------------------

n=2000              #No of samples
U1=np.random.rand(1,n)
U2=np.random.rand(1,n)

#Defining the terms to be used in Box-Muller
sq_term = np.sqrt(-2*np.log(U1))
cos_term = np.cos(2*np.pi*U2)
sin_term = np.sin(2*np.pi*U2)

N1 = sq_term*cos_term
N2 = sq_term*sin_term

#Transform 
mu = 0.2             #mean
sigma = 0.1          #standard deviation
N = N1*sigma + mu    # could also use N = N2*sigma+mu

#Defining the parameters
a = -0.2             #left truncation
b = 0.6              #right truncation
ngrid = 200          #number of grid intervals

dx = (b-a)/ngrid     #grid step
x = np.arange(a,b,dx)

#Computing the PDF
phi = ((x-mu)/sigma)**2
R = np.exp((-0.5)*phi)
f = (1/(np.sqrt(2*np.pi)*sigma))*R

phi = ((N[0]-mu)/sigma)**2
R = np.exp((-0.5)*phi)
fN = (1/(np.sqrt(2*np.pi)*sigma))*R

U =np.random.rand(1,n)
FN = U*fN

#Create Figure 
fig = plt.figure(figsize=(7,4), dpi=100)

#Add set of axes to figure
ax = fig.add_axes([0,0,1,1]) # left,bottom,width,height 

#Figure 1
#Plot on that set of axes
ax.plot(x,f, 'r', label='Analytical PDF')
plt.plot(N[0],FN[0],'b.', label='Sampled')
plt.title('Normal Random Numbers using Box Muller')
plt.xlim(a,b)
plt.legend()
plt.show()
```
