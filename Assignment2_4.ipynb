```python
#!/usr/bin/python
#Assignment2_4.ipynb
#Parth Parakh
#purpose: Formative Assignment 2_4

#---------------------------------------------------------------------
#import modules
import numpy as np
import matplotlib.pyplot as plt
#---------------------------------------------------------------------
# Accept reject method for normal
# Normal from the Laplce

# Calculate random values
n=1000000;
U1=np.random.rand(1,n);
U2=np.random.rand(1,n);
c=np.sqrt(np.exp(1)*2/np.pi);

# Calculate x generated as double exponential
x=(np.log(2*U1)*(U1<0.5))-(np.log(2*(1-U1))*(U1>0.5));
gx=0.5*np.exp(-np.abs(x));
fx=np.sqrt(1/(2*np.pi))*np.exp(-(x**2)/2);
           
# Save x according to the accept/reject algorithm
output=x*(U2*c<(fx/gx));

# Extract non zero values from output
output2=output[output!=0];

h,x=np.histogram(output2,100);
h2=h/(len(output2)*(x[1]-x[0]));

plt.plot(x[:-1],h2)
plt.xlabel('x')
plt.title('Standard normal distribution using the accept-reject algorithm')
plt.show()
```
