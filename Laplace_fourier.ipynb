```python
#!/usr/bin/python3
#Laplace_fourier.ipynb
#Parth Parakh
#purpose: # Check numerically the Fourier pair Laplace <-> Lorentzian
#---------------------------------------------------------------------
#import modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy import fft
from scipy import interpolate
from scipy.interpolate import CubicSpline
#----------------------------------------------------------------------

N = 1024;
Dt = 0.1;
t = Dt*np.arange(-N/2,N/2-1);
T = N*Dt;
# Nyquist relation: Dt*Dw = 2*pi/N
Dw = (2*np.pi)/T;
w = Dw*np.arange(-N/2,N/2-1);
W = N*Dw; # = 2*pi/Dt

a = 1;
fa = (a/2)*np.exp(-a*abs(t)); # Laplace (or double exponential)
Fa = a**2/(a**2+w**2); # Lorentzian (or Cauchy)

Fn  = fft.fftshift(fft.ifft(fft.ifftshift(fa)))*T;
fn  = fft.fftshift(fft.fft(fft.ifftshift(Fa)))/T;
Fn1 = fft.fftshift(fft.fft(fft.ifftshift(fa)))*Dt;
fn1 = fft.fftshift(fft.ifft(fft.ifftshift(Fa)))/Dt;

plot1 = plt.figure(1,figsize=(7,4), dpi=100)
plt.plot(t,np.real(fn),'r',label='Re(fn)')
plt.plot(t,np.imag(fn),'g', label='Im(fn)')
plt.plot(t,fa,'k:',label='fa')
plt.axis([-10,10,0,1])
plt.xlabel('t')
plt.ylabel('f')
plt.legend()

plot2 = plt.figure(2,figsize=(7,4), dpi=100)
plt.plot(w,np.real(Fn),'b',label='Re(Fn)')
plt.plot(w,np.imag(Fn),'m', label='Im(Fn)')
plt.plot(w,Fa,'c:',label='Fa')
plt.axis([-20,20,0,1])
plt.xlabel('omega')
plt.ylabel('F')
plt.legend()

plot3 = plt.figure(3,figsize=(7,4), dpi=100)
plt.plot(t,np.real(fn1),'r',label='Re(fn1)')
plt.plot(t,np.imag(fn1),'g', label='Im(fn1)')
plt.plot(t,fa,'k:',label='fa')
plt.axis([-10,10,0,1])
plt.xlabel('t')
plt.ylabel('f')
plt.legend()

plot4 = plt.figure(4,figsize=(7,4), dpi=100)
plt.plot(w,np.real(Fn),'b',label='Re(Fn1)')
plt.plot(w,np.imag(Fn),'m', label='Im(Fn1)')
plt.plot(w,Fa,'c:',label='Fa')
plt.axis([-20,20,0,1])
plt.xlabel('omega')
plt.ylabel('F')
plt.legend()
```
